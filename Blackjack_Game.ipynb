```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
```

```python
class Card:

    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                self.deck.append(created_card)
    
    def __str__(self):
        deck_str = ""
        for deck_of_card in self.deck:
            deck_str += str(deck_of_card) + "\n"
        return deck_str

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        return self.deck.pop()
```

```python
class Hand:
    def __init__(self):
        self.cards = []  
        self.total_card_value = 0
    
    def add_card(self,card):
        if type(card) == type([]):
            # list of multiple card object
            self.cards.extend(card)
        else:
            # For a single card object
            self.cards.append(card)
            
    def total_card_values(self):
        for card_in_hand in self.cards:
            self.total_card_value += card_in_hand.value
        return self.total_card_value
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100  
        self.bet = 0
        
    def win_bet(self):
        if total_card_value < 21:
            return True
    
    def lose_bet(self):
        if total_card_value > 21:
            return True
```

```python
def hit(deck,hand):
    
    hand.add_card(deck.deal())
    
```

```python
def hit_or_stand(deck,hand):
    
    player_choice = ' '
    
    while player_choice not in ['y','n']:
        player_choice = input("Player, Do you want to hit? (y or n): ")
        if player_choice not in ['y','n']:
            print("Invalid input, please Choose (y or n)")
    
    if player_choice == 'y':
        print("Player Hit")
        print("After player Hit")
        hit(deck,hand)
    else:
        print("player No Hit")
        return player_choice
```

```python
def show_some(player,dealer):
        
    for i in range (len(player.cards)):
        print(f"Player Card {i+1}: {player.cards[i]}")
        
    for i in range (len(dealer.cards)-1):
        print(f"Dealer Card {i+1}: {dealer.cards[i]}")
    
def show_all(player,dealer):
    
    for i in range (len(player.cards)):
        print(f"Player Card {i+1}: {player.cards[i]}")
        
    for i in range (len(dealer.cards)):
        print(f"Dealer Card {i+1}: {dealer.cards[i]}")
    
def show_some_value(player,dealer):
    return player.total_card_value

def show_all_value(player,dealer):
    return dealer.total_card_value
```

```python
def player_busts():
    if player.total_card_value > 21:
        print("Player Busts, Dealer Wins!")
        return True
    else:
        print("Player not Busts")

def dealer_busts():
    if dealer.total_card_value > 21:
        print("Dealer Busts, Player Wins!")
        return True
    else:
        print("Dealer not Busts")

def wins():
    if 21 >= player.total_card_value > dealer.total_card_value:
        print("Player won")
    elif 21 >= dealer.total_card_value > player.total_card_value:
        print("Dealer won")
    elif player.total_card_value == dealer.total_card_value:
        print("Tie")
```
